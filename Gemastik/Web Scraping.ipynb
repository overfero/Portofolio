```python
links = []
y = 1
kondisi = True
while(kondisi):
    req = requests.get(f"https://news.detik.com/berita/indeks/{y}").text
    soup = bs4.BeautifulSoup(req,'html.parser')
    url = soup.find_all(name='a',attrs={'class':'media__link'})
    for x in range(len(url)):
        links.append(url[x].get('href'))
    if len(links) > 4000:
        kondisi = False
    print(y)
    y+=1
links = pd.Series(links).unique()
links.shape,links
```

```python
header = []
body = []
y = 1
for x in links:
    req = requests.get(x).text
    soup = bs4.BeautifulSoup(req,'html.parser')
    head = soup.find(name='h1', attrs={'class':'detail__title'}).text
    header.append(head)
    bod = soup.find(name='div', attrs={'class':'detail__body-text itp_bodycontent'}).text
    body.append(bod)
    print(y)
    y+=1
```

```python
pd.Series(body).to_csv('fact_detik.csv', sep=',',index=False)
```
